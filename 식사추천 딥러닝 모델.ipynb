```python
import pandas as pd
import numpy as np
```

```python
data_ori = pd.read_csv('/filename.csv')

data_ori
```

```python
data = data_ori.iloc[:, 1:]

data
```

```python
data_X = data.iloc[:, :-1]
data_y = data.iloc[:,-1:]
```

```python
X_columns = ['나이', '성별', '시간', '기분', '날씨']

data_X.columns = X_columns
```

```python
data_X
```

```python
X = pd.get_dummies(data_X)

X
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
```

```python
y = le.fit_transform(data_y)
```

```python
le.classes_
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Input, Conv2D, Dense, Flatten, Dropout, GlobalMaxPooling2D, MaxPooling2D, BatchNormalization
from tensorflow.keras.models import Model
from tensorflow.keras import Sequential

from sklearn.metrics import classification_report, confusion_matrix
from tensorflow.keras.preprocessing import image
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y)
```

```python
model4 = Sequential()

model4.add(Dense(400, input_dim=15, activation='relu'))

model4.add(Dense(300, activation='relu'))
model4.add(Dense(200, activation='relu'))
model4.add(Dense(150, activation='relu'))
model4.add(Dense(100, activation='relu'))
model4.add(Dense(50, activation='relu'))

model4.add(Dense(3, activation='softmax'))

model4.summary()
```

```python
model4.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
r4 = model4.fit(X_train, y_train, epochs=100, validation_data=(X_test, y_test))
```

```python
input_ = [[0, 1, 0, 0, 1, 0, 0, 0, 1, 0, 0, 1, 0, 0, 1]]

prediction = model4.predict(input_)
print(prediction)
```
